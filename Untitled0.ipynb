### Entraînement d'un modèle d'apprentissage automatique simple

Tout d'abord, importons les bibliothèques nécessaires. Nous aurons besoin de `sklearn` pour le jeu de données, le modèle et les utilitaires.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Ensuite, nous allons charger le jeu de données Iris, qui est un jeu de données classique pour les tâches de classification. Il contient des mesures pour 150 fleurs d'iris, chacune appartenant à l'une des trois espèces.

```python
iris = load_iris()
X = iris.data
y = iris.target

# Afficher les 5 premières lignes des caractéristiques et des noms de cible
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]
display(df.head())
```

Maintenant, nous allons diviser nos données en ensembles d'entraînement et de test. Cela nous permet d'entraîner le modèle sur une partie des données et d'évaluer ses performances sur des données non vues.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Nous utiliserons un modèle de régression logistique simple. Nous l'initialisons, puis nous l'entraînons à l'aide de nos données d'entraînement.

```python
model = LogisticRegression(max_iter=200) # max_iter augmenté pour la convergence
model.fit(X_train, y_train)
print("Entraînement du modèle terminé !")
```

Enfin, nous faisons des prédictions sur l'ensemble de test et évaluons la précision du modèle.

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle sur l'ensemble de test : {accuracy:.2f}")
```
